```python
# dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
data = pd.read_csv('Alcohol_Consumption_US.csv')
data
```

```python
lg = sns.lineplot(data=data, x='Year', y='All beverages (Per capita consumption)', hue='State_abbrev', style='State_abbrev')


plt.figure(figsize=(12,8))

# adjusting the size and location of the legend
lg.legend(fontsize=8, bbox_to_anchor= (1.2,1.05))

# setting the title of the chart
lg.set_title('Total Alcohol Consumption of Each State Over Time')
```
